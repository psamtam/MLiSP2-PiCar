```python
def decode_img(
    img,
):  # -> Any | defaultdict | Any | list | None | object | Tensor | ...:
    # Convert the compressed string to a 3D uint8 tensor
    img = tf.io.decode_png(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    return img

def process_path(file_path, angle):
    # Load the raw data from the file as a string
    img = tf.io.read_file(file_path)
    img = decode_img(img)
    return img, angle

def image_id_to_path(image_id, angle):
    image_path = tf.strings.join([training_data_dir, '/', tf.as_string(image_id), ".png"])  # Assuming images are .jpg
    return image_path, angle
```

```python
def visualise(original, augmented):
    fig = plt.figure()
    plt.subplot(1, 2, 1)
    plt.title("Original image")
    plt.imshow(original.numpy().astype("float32"))

    plt.subplot(1, 2, 2)
    plt.title("Augmented image")
    plt.imshow(augmented.numpy().astype("float32"))
```

```python
rotate_by_degree = 5
random_rotate = keras.layers.RandomRotation(rotate_by_degree / 360)


def augment(image, angle, seed=None):
    # Deal with the random seed
    if seed == None:
        seed = tf.random.uniform(shape=[2], maxval=10000, dtype=tf.int32)
    else:
        seed = tf.constant([seed, seed], dtype=tf.int32)

    image = tf.image.stateless_random_brightness(image, 0.3, seed)
    image = tf.image.stateless_random_contrast(image, 0.75, 1.25, seed)
    image = tf.image.stateless_random_hue(image, 0.05, seed)
    image = tf.image.stateless_random_saturation(image, 0.7, 1.2, seed)

    image = random_rotate(image)

    # crop image
    image = tf.image.stateless_random_crop(image, size=[210, 280, 3], seed=seed)

    # resize image back to 240x320
    image = tf.image.resize(image, [240, 320])

    return image, angle
```

```python
def rephrase_ds(image, angle):
    return image, {"angle_output": angle}


train_ds = train_ds.map(rephrase_ds, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.map(rephrase_ds, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
batch_size = 64


def configure_for_performance(ds, shuffle: bool, batch: bool):
    if shuffle:
        ds = ds.shuffle(buffer_size=5000)
    if batch:
        ds = ds.batch(batch_size, drop_remainder=True)
    else:
        ds = ds.batch(1, drop_remainder=False)
    ds = ds.prefetch(buffer_size=tf.data.AUTOTUNE)
    return ds


train_ds = train_ds.repeat()
val_ds = val_ds.repeat()
train_ds = configure_for_performance(train_ds, shuffle=True, batch=True)
val_ds = configure_for_performance(val_ds, shuffle=False, batch=True)
```

### Build model

```python
image_shape = (240, 320, 3)

base_model = keras.applications.EfficientNetV2S(
    include_top=False,
    weights="imagenet",
    input_shape=image_shape,
    pooling=None,
    name="efficientnetv2-s",
)
base_model.trainable = False
```

```python
feature_extractor = base_model.output

# angle

# path 1
angle_global_features = layers.GlobalAveragePooling2D(name='angle_global_pool')(feature_extractor)
angle_global_features = layers.BatchNormalization(name='angle_global_bn')(angle_global_features)
angle_global_features = layers.Dropout(0.2, name='angle_global_dropout')(angle_global_features)

# Path 2: Enhanced Convolutional Branch
angle_conv_1 = layers.Conv2D(128, (3, 3), padding="same", activation="relu", name="angle_conv_1")(feature_extractor)
angle_conv_1 = layers.BatchNormalization(name="angle_conv_1_bn")(angle_conv_1)

angle_conv_2 = layers.Conv2D(256, (3, 3), padding="same", activation="relu", name="angle_conv_2")(angle_conv_1)
angle_conv_2 = layers.BatchNormalization(name="angle_conv_2_bn")(angle_conv_2)

angle_conv_3 = layers.Conv2D(128, (3, 3), padding="same", activation="relu", name="angle_conv_3")(angle_conv_2)
angle_conv_3 = layers.BatchNormalization(name="angle_conv_3_bn")(angle_conv_3)

# Residual connection
angle_conv_residual = layers.Conv2D(128, (1, 1), padding="same", name="angle_conv_residual")(feature_extractor)
angle_conv_features = layers.Add(name="angle_conv_add")([angle_conv_3, angle_conv_residual])
angle_conv_features = layers.Flatten(name="angle_conv_flatten")(angle_conv_features)

# Path 3: Attention Branch
angle_attention = layers.GlobalAveragePooling2D(name="angle_attention_pool")(feature_extractor)
angle_attention = layers.Dense(1280 // 16, activation="relu", name="angle_attention_reduce")(angle_attention)  # 1280 = EfficientNetV2S output channels
angle_attention = layers.Dense(1280, activation="sigmoid", name="angle_attention_expand")(angle_attention)
angle_attention = layers.Reshape((1, 1, 1280), name="angle_attention_reshape")(angle_attention)
angle_attention_features = layers.Multiply(name="angle_attention_multiply")([feature_extractor, angle_attention])
angle_attention_features = layers.GlobalAveragePooling2D(name="angle_attention_global_pool")(angle_attention_features)

# Merge all three paths
angle_merged_features = layers.Concatenate(name="angle_merge")(
    [angle_global_features, angle_conv_features, angle_attention_features]
)
angle_merged_features = layers.BatchNormalization(name='angle_merge_bn')(angle_merged_features)

angle_hidden_5 = layers.Dense(2560, activation="relu", kernel_initializer="he_uniform", name="angle_hidden_5")(angle_merged_features)
angle_hidden_5 = layers.Dropout(0.3, name='angle_hidden_5_dropout')(angle_hidden_5)
angle_hidden_4 = layers.Dense(2560, activation="relu", kernel_initializer="he_uniform", name="angle_hidden_4")(angle_hidden_5)
angle_hidden_4 = layers.Dropout(0.3, name='angle_hidden_4_dropout')(angle_hidden_4)
angle_hidden_3 = layers.Dense(1024, activation="relu", kernel_initializer="he_uniform", name="angle_hidden_3")(angle_hidden_4)
angle_hidden_2 = layers.Dense(512, activation="relu", kernel_initializer="he_uniform", name="angle_hidden_2")(angle_hidden_3)
angle_hidden = layers.Dense(100, activation="relu", kernel_initializer="he_uniform", name="angle_hidden")(angle_hidden_2)
angle_output = layers.Dense(1, activation="linear", name="angle_output")(angle_hidden)

model = Model(inputs=base_model.input, outputs=[angle_output], name="my_model")

learning_rate = 0.004

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
    loss={"angle_output": "mse"},
    metrics={"angle_output": ["mae"]},
)

# model.summary()
```

```python
def lr_schedule(epoch, lr):
    initial_lr = learning_rate
    decay = 0.6
    return initial_lr / (1 + ((epoch + 1) // 2) * decay)

# for epoch in range(30):
#     print(f"Epoch {epoch+1:2d}: {lr_schedule(epoch, 0):.6f}")
```

```python
checkpoint = keras.callbacks.ModelCheckpoint(
    filepath="checkpoints/model_epoch_{epoch:02d}.keras",
    save_weights_only=False,
    save_freq="epoch",
    verbose=1,
)

lr_scheduler = keras.callbacks.LearningRateScheduler(lr_schedule, verbose=1)

history_callback = keras.callbacks.History()

try:
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=30,
        steps_per_epoch= train_ds_length // batch_size,
        validation_steps= val_ds_length // batch_size,
        callbacks=[checkpoint, history_callback, lr_scheduler],
        verbose=1,
    )
except KeyboardInterrupt:
    print("Stopped early—saving history")
    with open("training_history.pkl", "wb") as f:
        pickle.dump(history_callback.history, f)
```

```python
timestamp = datetime.now().strftime("%Y%m%d_%H%M")  # Format: YYYYMMDD_HHMM
name_template = f'v2_angle_model_{timestamp}'

folder_path = f"Models/{name_template}"

if not os.path.exists(folder_path):
    os.makedirs(folder_path)

filename = f"{folder_path}/{name_template}.keras"
model.save(filename)
filename = f"{folder_path}/{name_template}.weights.h5"
model.save_weights(filename)

history = history_callback.history

filename = f"{folder_path}/{name_template}_history.pkl"
with open(filename, 'wb') as f:
    pickle.dump(history, f)
```
