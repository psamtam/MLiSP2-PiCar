```python
import numpy as np
import pandas as pd
import os
from datetime import datetime
```

```python
old_csv_dir = "/mnt/c/Users/psamt/OneDrive - The University of Nottingham/_Spring/PHYS4036_Machine Learning in Science Part II/Project/Captured_data/processed/T_right"

old_csv_path = os.path.join(old_csv_dir, "_labels_old.csv")

old_csv_df = pd.read_csv(old_csv_path).set_index("image_id")

old_csv_df
```

```python
png_image_ids = [f.split('.')[0] for f in os.listdir(old_csv_dir) if f.endswith(".png")]
```

```python
new_csv_df = old_csv_df[old_csv_df.index.isin(png_image_ids)]

timestamp = datetime.now().strftime("%Y%m%d_%H%M")  # Format: YYYYMMDD_HHMM

new_csv_df.to_csv(os.path.join(old_csv_dir, f"_labels_{timestamp}.csv"))
```
