```python
X_train_path = [
    os.path.join(training_data_dir, str(image_id) + ".png") for image_id in X_train
]
X_val_path = [
    os.path.join(training_data_dir, str(image_id) + ".png") for image_id in X_val
]
```

```python
def decode_img(
    img,
):  # -> Any | defaultdict | Any | list | None | object | Tensor | ...:
    # Convert the compressed string to a 3D uint8 tensor
    img = tf.io.decode_png(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    return img

def process_path(file_path, speed):
    # Load the raw data from the file as a string
    img = tf.io.read_file(file_path)
    img = decode_img(img)
    return img, speed

def image_id_to_path(image_id, speed):
    image_path = tf.strings.join([training_data_dir, '/', tf.as_string(image_id), ".png"])  # Assuming images are .jpg
    return image_path, speed
```

```python

train_ds = tf.data.Dataset.from_tensor_slices((X_train_path, y_train))
val_ds = tf.data.Dataset.from_tensor_slices((X_val_path, y_val))

train_ds = train_ds.map(process_path, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.map(process_path, num_parallel_calls=tf.data.AUTOTUNE)

train_ds_length = tf.data.experimental.cardinality(train_ds).numpy()
print(train_ds_length)
val_ds_length = tf.data.experimental.cardinality(val_ds).numpy()
print(val_ds_length)
```

```python
def visualise(original, augmented):
    fig = plt.figure()
    plt.subplot(1, 2, 1)
    plt.title("Original image")
    plt.imshow(original.numpy().astype("float32"))

    plt.subplot(1, 2, 2)
    plt.title("Augmented image")
    plt.imshow(augmented.numpy().astype("float32"))
```

```python
rotate_by_degree = 5
random_rotate = keras.layers.RandomRotation(rotate_by_degree / 360)


def augment(image, speed, seed=None):
    # Deal with the random seed
    if seed == None:
        seed = tf.random.uniform(shape=[2], maxval=10000, dtype=tf.int32)
    else:
        seed = tf.constant([seed, seed], dtype=tf.int32)

    image = tf.image.stateless_random_brightness(image, 0.3, seed)
    image = tf.image.stateless_random_contrast(image, 0.75, 1.25, seed)
    image = tf.image.stateless_random_hue(image, 0.05, seed)
    image = tf.image.stateless_random_saturation(image, 0.7, 1.2, seed)
    image = tf.image.stateless_random_jpeg_quality(image, 80, 100, seed)

    image = random_rotate(image)

    # crop image
    image = tf.image.stateless_random_crop(image, size=[210, 280, 3], seed=seed)

    # resize image back to 240x320
    image = tf.image.resize(image, [240, 320])

    return image, speed
```

```python
def scale_ds(image, speed):
    # Range of input image: [0, 1]
    return image * 2 - 1.0, speed


train_ds = train_ds.map(scale_ds, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.map(scale_ds, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
def rephrase_ds(image, speed):
    return image, {"speed1_output": speed, "speed2_output": speed, "speed3_output": speed, "speed4_output": speed,  "speed5_output": speed}
    # return image, {"speed1_output": speed, "speed2_output": speed, "speed3_output": speed}


train_ds = train_ds.map(rephrase_ds, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.map(rephrase_ds, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
batch_size = 64


def configure_for_performance(ds, shuffle: bool, batch: bool):
    if shuffle:
        ds = ds.shuffle(buffer_size=3000)
    if batch:
        ds = ds.batch(batch_size, drop_remainder=True)
    else:
        ds = ds.batch(1, drop_remainder=False)
    ds = ds.prefetch(buffer_size=tf.data.AUTOTUNE)
    return ds


train_ds = train_ds.repeat()
val_ds = val_ds.repeat()
train_ds = configure_for_performance(train_ds, shuffle=True, batch=True)
val_ds = configure_for_performance(val_ds, shuffle=False, batch=True)
```

### Build model

```python
print(y_train)
class_weight = compute_class_weight("balanced", classes=np.array([0, 1]), y=y_train)
class_weight_dict = {0: class_weight[0], 1: class_weight[1]}
print(class_weight_dict)
```

```python
image_shape = (240, 320, 3)

base_model = keras.applications.MobileNetV3Large(
    include_top=False,
    weights="imagenet",
    input_shape=image_shape,
    pooling=None,
    include_preprocessing=False, 
)
base_model.trainable = False

print(base_model.output.shape)
```

```python
from os import name


feature_extractor = base_model.output

print(feature_extractor.shape)

s = feature_extractor
s_attention = layers.Conv2D(1, (1, 1), padding='same', activation='sigmoid', name='s1_attention_conv')(s)
s = layers.Multiply(name='s1_attention_apply')([s, s_attention])
s = layers.GlobalAveragePooling2D(name='s1_attention_GAP')(s)
s = layers.Dropout(0.3, name='s1_dropout_2')(s)
s = layers.Dense(128, activation='relu', kernel_initializer='he_uniform', name='s1_dense_1')(s)
s = layers.Dropout(0.3, name='s1_dropout')(s)
s = layers.Dense(1, activation='sigmoid', kernel_initializer='he_uniform', name='s1_output')(s)

s1 = s


s = layers.Conv2D(512, (3, 3), (2, 2), padding='same', activation='relu', name='s2_conv_1')(feature_extractor)
s_attention = layers.Conv2D(1, (1, 1), padding='same', activation='sigmoid', name='s2_attention_conv')(s)
s = layers.Multiply(name='s2_attention_apply')([s, s_attention])
s = layers.GlobalAveragePooling2D(name='s2_attention_GAP')(s)
s = layers.Dropout(0.3, name='s2_dropout')(s)
s = layers.Dense(72, activation='relu', kernel_initializer='he_uniform', name='s2_dense_1')(s)
s = layers.Dense(1, activation='sigmoid', kernel_initializer='he_uniform', name='s2_output')(s)

s2 = s


s = layers.Conv2D(256, (3, 3), (2, 2), padding='same', activation='relu', name='s3_conv_1')(feature_extractor)
s_attention = layers.Conv2D(1, (1, 1), padding='same', activation='sigmoid', name='s3_attention_conv')(s)
s = layers.Multiply(name='s3_attention_apply')([s, s_attention])
s = layers.GlobalAveragePooling2D(name='s3_attention_GAP')(s)
s = layers.Dropout(0.3, name='s3_dropout')(s)
s = layers.Dense(64, activation='relu', kernel_initializer='he_uniform', name='s3_dense_1')(s)
s = layers.Dense(1, activation='sigmoid', kernel_initializer='he_uniform', name='s3_output')(s)

s3 = s


s = layers.Conv2D(64, (3, 3), (2, 2), padding='same', activation='relu', name='s4_conv_1')(feature_extractor)
s_attention = layers.Conv2D(1, (1, 1), padding='same', activation='sigmoid', name='s4_attention_conv')(s)
s = layers.Multiply(name='s4_attention_apply')([s, s_attention])
s = layers.GlobalAveragePooling2D(name='s4_attention_flatten')(s)
s = layers.Dropout(0.2, name='s4_dropout_2')(s)
s = layers.Dense(64, activation='relu', kernel_initializer='he_uniform', name='s4_dense_1')(s)
s = layers.Dropout(0.2, name='s4_dropout')(s)
s = layers.Dense(1, activation='sigmoid', kernel_initializer='he_uniform', name='s4_output')(s)

s4 = s


s = layers.Conv2D(48, (3, 3), (3, 3), padding='same', activation='relu', name='s5_conv_1')(feature_extractor)
s = layers.Flatten(name='s5_flatten')(s)
s = layers.Dense(64, activation='relu', kernel_initializer='he_uniform', name='s5_dense_1')(s)
s = layers.Dropout(0.3, name='s5_dropout')(s)
s = layers.Dense(1, activation='sigmoid', kernel_initializer='he_uniform', name='s5_output')(s)

s5 = s

model = Model(inputs=base_model.input, outputs={'speed1_output': s1, 'speed2_output': s2, 'speed3_output': s3, 'speed4_output': s4, 'speed5_output': s5}, 
                                                name="speed_model")

learning_rate = 0.0010

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
    # loss={"speed1_output": WeightedMeanSquaredError(class_weight_dict)},
    loss={"speed1_output": WeightedBinaryCrossEntropy(class_weight_dict),
          "speed2_output": WeightedBinaryCrossEntropy(class_weight_dict),
          "speed3_output": WeightedBinaryCrossEntropy(class_weight_dict),
          "speed4_output": WeightedBinaryCrossEntropy(class_weight_dict),
          "speed5_output": WeightedBinaryCrossEntropy(class_weight_dict)},
    metrics={"speed1_output": ["mse", "accuracy"],
             "speed2_output": ["mse", "accuracy"],
             "speed3_output": ["mse", "accuracy"],
             "speed4_output": ["mse", "accuracy"],
             "speed5_output": ["mse", "accuracy"]},
)

# model.summary()
```

```python
def lr_schedule(epoch, lr):
    FIRST_N_EPOCHS = 2
    
    initial_lr = learning_rate
    if epoch < FIRST_N_EPOCHS:
        return max(initial_lr * 2, 0.002)
    decay = 0.3
    return initial_lr / (1 + ((epoch-FIRST_N_EPOCHS) // 3) * decay)

for epoch in range(30):
    print(f"Epoch {epoch+1:2d}: {lr_schedule(epoch, 0):.6f}")
```

### Train model

```python
checkpoint = keras.callbacks.ModelCheckpoint(
    filepath="checkpoints/model_epoch_{epoch:02d}.keras",
    save_weights_only=False,
    save_freq="epoch",
    verbose=1,
)

lr_scheduler = keras.callbacks.LearningRateScheduler(lr_schedule, verbose=1)

history_callback = keras.callbacks.History()

NUM_OF_EPOCHS = 50

try:
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=2,
        steps_per_epoch= train_ds_length // batch_size,
        validation_steps= val_ds_length // batch_size,
        callbacks=[checkpoint, history_callback, lr_scheduler],
        verbose=1,
    )
except KeyboardInterrupt:
    print("Stopped earlyâ€”saving history")
    with open("training_history.pkl", "wb") as f:
        pickle.dump(history_callback.history, f)
```
